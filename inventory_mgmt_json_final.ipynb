# Cafe Inventory Management System

```python
import time
import json


#read json inventory file
fd = open('Records.json','r')
js = fd.read()
fd.close()

#loaded the inventory into dictionary using loads
record = json.loads(js) 



# Show the menu to customer

print("------------------MENU-------------------\n")
print("Product ID", "|" , "Name", "|", "Price", "|" , "Quantity", "\n")
for key in record.keys():
    print(key , "      :" , record[key]['Name'] , "|" , record[key]['Price'] , "|" , record[key]['Qty'])

#take user input
ui_username = str(input("Enter your name  : "))
ui_phone = str(input("Enter your phone no : "))
ui_mail = str(input("Enter your mail id   : "))
ui_prod_id = (input("Enter Product ID     : "))
ui_qty     = int(input("Enter quantity    : "))

#if logic to check whether quantity in stock or not

if ui_qty <= record[ui_prod_id]['Qty']:
    #update inventory quantity record
    record[ui_prod_id]['Qty'] = record[ui_prod_id]['Qty'] - ui_qty 

    #print bill
    print("\n------------------BILL-------------------\n")

    print("Product Name   :", record[ui_prod_id]['Name'])
    print("Price          :", record[ui_prod_id]['Price'])
    print("Quantity       :", ui_qty)
    print("-------------------------------------------")
    bill1 = ui_qty * record[ui_prod_id]['Price']
    #check if discount eligible
    if bill1 > 2000:
        print("You are eligible for 5'%' discount\n")
        print("Bill Amount      :", bill1)
        print("Discount         :", 0.05 * bill1)
        print("Net Bill Amount  :", bill1 - 0.05*bill1)
        bill1 = bill1 - 0.05*bill1
    else:
        print("Bill Amount    :", bill1)
     

    #Generating sales report
    sale = ui_username +','+ ui_phone +','+ ui_mail +','+ui_prod_id +','+record[ui_prod_id]['Name'] +','+ str(ui_qty) +','+ str(record[ui_prod_id]['Price']) +','+ str(bill1) +','+ time.ctime() + '\n'
    


#if quantity asked by ucstomer not available
else:
    print("Sorry, we do not have enough stock.\nWe have only ", record[ui_prod_id]['Qty'], "\nWould you like to purchase?")
    cust_ip = input("Enter Y/N : ")
    if cust_ip == 'Y' or cust_ip == 'y':
        #print bill
        print("\n------------------BILL-------------------\n")

        print("Product Name   :", record[ui_prod_id]['Name'])
        print("Price          :", record[ui_prod_id]['Price'])
        print("Quantity       :", record[ui_prod_id]['Qty'])
        print("------------------------------------------")
        bill2 = record[ui_prod_id]['Qty'] * record[ui_prod_id]['Price']
        #check if eligible for discount
        if bill2 > 2000:
            print("You are eligible for 5'%' discount\n")
            print("Bill Amount      :", bill2)
            print("Discount         :", 0.05 * bill2)
            print("Net Bill Amount  :", bill2 - 0.05*bill2)
            bill2 = bill2 - 0.05*bill2
        else:
            print("Bill Amount    :", bill2)

        #Generating sales report
        sale = ui_username +','+ ui_phone +','+ ui_mail +','+ui_prod_id +','+record[ui_prod_id]['Name'] +','+ str(record[ui_prod_id]['Qty']) +','+ str(record[ui_prod_id]['Price']) +','+ str(bill2) +','+ time.ctime() + '\n'
        
        #updating qty stock to 0
        record[ui_prod_id]['Qty'] = 0

        

    else:
        print("Thank you, come again.")


#updating sales txt file
fd = open('sales2.txt','a')
fd.write(sale)
fd.close()
        
#record into json format
js = json.dumps(record)
#write into a json file
fd = open("Records.json",'w')       #since we are overwriting everytime
fd.write(js)
fd.close()



```
